```python
import redis
```

```python
r = redis.StrictRedis()
```

```python
for blob in blobs:
    pipe = r.pipeline()
    
    path, name = os.path.split(blob.name)
    parts = (part + '/' for part in path.split('/') if part != '')
    base = '/'
    
    for part in parts:
        pipe.sadd(base, part)
        #dirs[base].add(part)
    
        base += part
    #dirs[base].add(name)
    pipe.sadd(base, name)
    
    name = base + name
    
    # set times and size
    pipe.hset(name, 'st_mtime', blob.updated.strftime('%s'))
    pipe.hset(name, 'st_atime', blob.updated.strftime('%s'))
    pipe.hset(name, 'st_ctime', blob.time_created.strftime('%s'))
    
    pipe.execute()
```
